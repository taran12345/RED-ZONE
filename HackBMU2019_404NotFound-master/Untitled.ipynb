```python
def zone(x):
    if (x["mean_crime"] >= 0) & (x["mean_crime"]<100):
        return 0
    elif (x["mean_crime"] >= 100) & (x["mean_crime"]<200):
        return 1
    else:
        return 2

```

```python
crime = crime.assign(alert_zone=crime.apply(zone, axis=1))
```
